```python
import pandas as pd
import numpy as np
```

```python
df_chicago = pd.read_csv("listings (2).csv")
df_new_orleans = pd.read_csv("listings (3).csv")
```

```python
# Fill missing values
df_chicago['reviews_per_month'] = df_chicago['reviews_per_month'].fillna(0)
```

```python

# Drop important nulls
df_chicago = df_chicago.dropna(subset=['price', 'neighbourhood', 'room_type'])

```

```python

# Clean price
df_chicago['price'] = df_chicago['price'].replace(r'[\$,]', '', regex=True)

```

```python

# Convert date
df_chicago['last_review'] = pd.to_datetime(df_chicago['last_review'], errors='coerce')

```

```python


# Remove duplicates
df_chicago = df_chicago.drop_duplicates()

```

```python


# Fix text
df_chicago['neighbourhood'] = df_chicago['neighbourhood'].str.lower().str.strip()
df_chicago['room_type'] = df_chicago['room_type'].str.lower().str.strip()

```

```python

# Remove outliers
df_chicago = df_chicago[df_chicago['price'] < 1000]
df_chicago = df_chicago[df_chicago['minimum_nights'] < 365]

```

```python

# Add city column
df_chicago['city'] = 'Chicago'
```

```python
df_chicago
```

```python
# Fill missing values
df_new_orleans['reviews_per_month'] = df_new_orleans['reviews_per_month'].fillna(0)
```

```python

# Drop important nulls
df_new_orleans = df_new_orleans.dropna(subset=['price', 'neighbourhood', 'room_type'])

```

```python

df_new_orleans['price'] = df_new_orleans['price'].replace(r'[\$,]', '', regex=True)
```

```python


# Convert date
df_new_orleans['last_review'] = pd.to_datetime(df_new_orleans['last_review'], errors='coerce')

```

```python


# Remove duplicates
df_new_orleans = df_new_orleans.drop_duplicates()

```

```python

# Fix text
df_new_orleans['neighbourhood'] = df_new_orleans['neighbourhood'].str.lower().str.strip()
df_new_orleans['room_type'] = df_new_orleans['room_type'].str.lower().str.strip()

```

```python

# Remove outliers
df_new_orleans = df_new_orleans[df_new_orleans['price'] < 1000]
df_new_orleans = df_new_orleans[df_new_orleans['minimum_nights'] < 365]
```

```python

# Add city column
df_new_orleans['city'] = 'New Orleans'
```

```python
df_new_orleans
```

```python
# combine both
df = pd.concat([df_chicago, df_new_orleans], ignore_index=True)
```
